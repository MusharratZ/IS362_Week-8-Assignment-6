```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

#1. Load data 
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
column_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'car_name']
auto_data = pd.read_csv(url, delim_whitespace=True, names=column_names)

```

```python
#2 Reveiw the data and set description
print(auto_data.head())
print(auto_data.info())
```

```python
# 3. Columns are named in step 1
# 4. Replace '?' in horsepower
auto_data['horsepower'] = pd.to_numeric(auto_data['horsepower'], errors='coerce')
```

```python
#5 convert origin column values to USA, Asia and Europe
auto_data['origin'] = auto_data['origin'].map({1: 'USA', 2: 'Asia', 3: 'Europe'})
```

```python
#6. Create a bar chart for distribution for cylinders
plt.figure(figsize=(8, 6))
sns.countplot(x='cylinders', data=auto_data)
plt.title('Distribution of Cylinders')
plt.xlabel('Cylinders')
plt.ylabel('Frequency')
plt.show()
```

```python
# 7 Create a Scatterplot showing relatnshp bwteen horsepower and weight
plt.figure(figsize=(8, 6))
sns.scatterplot(x='horsepower', y='weight', data=auto_data)
plt.title('Relationship between Horsepower and Weight')
plt.xlabel("Origin")
plt.ylabel('Weight')
plt.show()
```

```python
# 8. Visualizing the relationship between mpg and origin
plt.figure(figsize=(8, 6))
sns.boxplot(x='origin', y='mpg', data=auto_data)
plt.title('Distribution of MPG by Origin')
plt.xlabel('Origin')
plt.ylabel('MPG')
plt.show()
```
